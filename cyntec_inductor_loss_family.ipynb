### Input Parameters

```python
input_params = {'vin' : 28,
                'vout': 9,
                'pin' : 165, 
                'eff' : .97, 
                'fs'  :300e3,
                'ton_mult':1,
                'tambient':45,
                'lout':{'family'   :'hbed053t',
                        'value(uH)':0.68,
                        'config'   :'series'}, #single, series, parallel
                'lvl_config':'3 level',
               }
input_params['iout'] = input_params['pin']*input_params['eff']/input_params['vout']
```

```python

lvalue_min = 0.1
lvalue_max = 3.3
step = 0.01
lvalue_list = np.arange(lvalue_min,lvalue_max+step,step).round(2).tolist()

wb=openpyxl.load_workbook(filename=cyntec_filename)
l_families = wb.sheetnames  # can use a subset list
```

### Inductor List Results

```python
l_result = pd.concat([l_set(input_params,family,lvalue_list) for family in l_families])
```
